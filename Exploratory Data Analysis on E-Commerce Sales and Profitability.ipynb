# 4> Feature engineering

```python
# Using the shipping date or order date make the one new columns Delivery_days
df["Delovery_Days"] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
# Cerate Month, Year from Order_Date
df["order Month"] = df["Order Date"].dt.strftime('%b')
df["order year"] = df["Order Date"].dt.year
df["month year"] = df["Order Date"].dt.strftime('%b-%Y')
```

```python
# Make a profit Ratio columns
df["Profit Ratio"] = df['Profit'] / df['Sales']
```

# 5>EDA

# 1>Univariate Analysis (Single Column)

```python
Category1_columns= [
 'Ship Mode',
 'State',
 'Delovery_Days']
```

```python
# Take category columns and cerate a count plot 
for col in Category1_columns:
    plt.figure(figsize=(10,5))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Count Plot for {col}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

# Count Plot Insights

## 1>Count Plot for Ship Mode Insights
- The Standard Class shipping mode is the most commonly used, with around 6000 orders.
- Second Class and First Class follow with approximately 2000 and 1600 orders, respectively.
- The Same Day shipping mode is the least used, with only about 600 orders.

## 2>Count Plot For State Insights
- California has the highest number of orders, with around 2000, followed by New York, Texas, and Washington.
- States like Wyoming, West Virginia, North Dakota, and South Dakota show minimal order counts.
- There is a steep drop in order volume after the top 5 states, indicating a high concentration of sales in specific regions.

                                                                           
## 3>Count plot for delivery_Days Insights
- Most orders are delivered in 4 to 5 days, with 4-day delivery being the most frequent (~2800 orders).
- Fewer deliveries occur in 0–2 days, and 6-day deliveries are also moderately common.
- Very fast delivery (0–1 days) is least common, indicating limited use of express shipping.

# 2>Bivariate Analysis 

# 1>Category vs Sales 
- Bar Plot
- (category + numeric)
- Helps businesses understand which product categories drive revenue.

```python
sns.barplot(data=df,x='Category',y='Sales', estimator=sum)# estimator=sum using this that show categoryvise total sales
plt.title('Total Sales by Category')
plt.show()
```

# Category vs Sales Insights

- Technology leads with the highest total sales, crossing ₹850,000.

- Furniture and Office Supplies have similar sales figures, both around ₹730,000–₹750,000.

# 2>Sub-Category vs Profit
- Bar Plot
- (category + numeric)
- That show which sub_category can gain high profit

```python
sns.barplot(x='Sub-Category',y='Profit',data=df,estimator=sum)
plt.xticks(rotation=90)
plt.title('Profit by Sub-Category')
plt.show()
```

# Sub-Category vs Profit insights
- The Copiers sub-category generates the highest profit, followed by Phones, Paper, and Accessories.
- Sub-categories like Chairs, Binders, Storage, and Appliances also show strong profitability.
- On the other hand, Tables and Bookcases are loss-making, with Tables showing the largest negative profit.

# 3>Discount vs Profit
- Scatter Plot
- (Numeric + Numeric)
- scatter plot to visually understand the relationship between discounts and profitability

```python
sns.scatterplot(x='Discount',y='Profit', data=df)
plt.title('Discount vs Profit')
plt.show()
```

# Discount vs Profit insights
- As the discount increases, profit tends to decrease.

- Higher discounts (above 0.3 or 30%) are mostly associated with negative profits.

- The highest profits are seen when the discount is near 0%, showing that full-price sales are most profitable.

- Even moderate discounts (10–20%) show reduced profitability compared to no discount.

# 4>Segment vs Profit

# 6>Order Month vs Sales
- Line plot
- (Category + numeric)
- We check the monthwise Sales

```python
monthly = df.groupby('order Month') ['Sales'].sum().reset_index() # It calculates the total sales for each month and stores the result in a new DataFrame called monthly.
sns.lineplot(x='order Month', y='Sales', data=monthly)
plt.title('Monthly Sales Trend')
plt.show()
```

# Order Month vs Sales Insights
- November recorded the highest sales, peaking at approximately ₹350,000—likely due to festive or holiday shopping.

- December and September also showed strong sales, both exceeding ₹300,000.

- A sharp drop in January and February, with February being the lowest month (~₹60,000).

- Gradual recovery starts from March onward, with a noticeable spike in June and July.

- May and October show moderate dips between high-performing months.

# 2>  Region + Sub-Category vs Profit
- Heat Map
-  Reveals profitable combinations 

```python
pivot = df.pivot_table(index='Region', columns='Sub-Category', values='Profit', aggfunc='sum') # Total Profit for each Sub-Category in each Region
sns.heatmap(pivot, cmap='coolwarm', annot=False, fmt=".0f")
plt.title('Profit by Region and Sub-Category')
plt.show()
```

# Heatmap Insights – Regional Profit vs. Sub-Category Performance
 ## 1>Top Performing Categories

- These can be considered "evergreen categories", providing a stable and reliable revenue stream for the company.

 ## 2>Loss-Making Categories
- Tables and Bookcases showed consistent losses in all regions, especially in East and South.
- These categories need immediate attention as they are clear indicators of underperformance.

 ## 3> Best Performing Region
- The East region showed positive profits across almost every sub-category, indicating:
- "Strong customer demand
- "Efficient operations
This makes East a high-potential market for future growth.

 ## 4>Underperforming Regions
- The Central and South regions showed frequent losses across multiple sub-categories.
- This suggests there may be market challenges or operational inefficiencies.
- A region-specific strategy update is needed to improve performance in these areas.

# 3>Discount + Sales + Profit Ratio
- Scatter Plot 
- It shows how discount, sales, and profit ratio are related

```python
sns.scatterplot(x='Discount', y='Sales', size='Profit Ratio', hue='Category', data=df)
plt.title('Discount vs Sales with Profit Ratio')
plt.show()
```

# Discount vs Sales with Profit Ratio – Graph Insights
## 1>Highest Sales & Profit at 0% Discount
- Products sold at full price (0% discount) achieved the highest sales and profit ratio.
Especially in the Technology category (green dots), many items performed well without any discount — showing strong demand and brand value.

## 2>Increasing Discounts Lead to Lower Profit
- As discount percentages increased, the profit ratio decreased, often turning negative.
This highlights that heavy discounting is not a sustainable strategy and can hurt overall profitability.

## 3>High Discounts Didn’t Boost Sales
- At 60% or more discount, sales volume was still very low.
- This means that deep discounts did not motivate customers, or the profit margins became too thin or negative, making the sales unprofitable.

## 4>Furniture Category Showed Mixed Trends
- The Furniture category (blue dots) showed some good sales even with discounts, but overall, the profit ratio remained low.
- This could be due to high shipping or handling costs, reducing net profit despite decent sales numbers.
